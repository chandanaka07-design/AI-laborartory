```python
import pandas as pd
```

```python
import json

with open('/content/Dataset.datatab', 'r') as f:
    json_data = json.load(f)

data = {}
for variable in json_data['variables']:
    col_name = variable['name']
    # The first element of dataRaw is the column name itself, so we skip it.
    col_data = variable['dataRaw'][1:]
    data[col_name] = col_data

df = pd.DataFrame(data)
display(df.head())
```

```python
print(df.info())
```

```python
# Convert 'Salary' to numeric, coercing errors to NaN
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Group by 'Company' and calculate the average 'Salary'
average_salary_by_company = df.groupby('Company')['Salary'].mean().reset_index()

# Display the result
display(average_salary_by_company)
```

```python
# Count the occurrences of each unique value in the 'Gender' column
gender_counts = df['Gender'].value_counts()

# Display the result
display(gender_counts)
```
