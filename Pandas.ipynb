### Load the Dataset
First, let's load the `Employee_dataset.csv` into a pandas DataFrame and display its initial information and the first few rows to understand its structure.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Employee_dataset.csv')

# Display basic information and the first few rows
print("Initial DataFrame Info:")
df.info()
print("\nInitial DataFrame Head:")
display(df.head())
```

### Handle Duplicate Records
We'll identify and remove any exact duplicate rows in the dataset.

```python
print(f"Number of rows before removing duplicates: {len(df)}")
df.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {len(df)}")
display(df.head())
```

### Handle Missing Values
We'll check for missing values, especially in the 'Salary' column, and decide on an appropriate imputation strategy (e.g., fill with median for numerical columns).

```python
print("Missing values before imputation:")
display(df.isnull().sum())

# Convert 'Salary' to numeric, coercing errors to NaN
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Impute missing salaries with the median salary
median_salary = df['Salary'].median()
df['Salary'].fillna(median_salary, inplace=True)

print("\nMissing values after imputation:")
display(df.isnull().sum())
display(df.head())
```

### Inconsistent Date Formats
We'll convert any date columns (e.g., 'Hire_Date') to a consistent datetime format. I'll assume 'Hire_Date' is the relevant date column and try to infer its format.

```python
# Assuming a 'Hire_Date' column exists and needs conversion
# If a different date column name exists, please update 'Hire_Date'
if 'Hire_Date' in df.columns:
    df['Hire_Date'] = pd.to_datetime(df['Hire_Date'], errors='coerce')
    print("\n'Hire_Date' column after conversion:")
    display(df[['Hire_Date']].head())
    print(f"Number of invalid dates in 'Hire_Date' after conversion: {df['Hire_Date'].isnull().sum()}")
else:
    print("No 'Hire_Date' column found for date format conversion.")
display(df.info())
```

### Erroneous Entries (Non-numeric salary values)
We already handled non-numeric salary values by coercing them to `NaN` and then imputing, so the 'Salary' column should now be clean and numeric.

```python
print("Checking 'Salary' column data type after cleaning:")
display(df['Salary'].dtype)
display(df['Salary'].describe())
```

### Compound Columns (e.g., 'Department_Region')
If there's a compound column like 'Department_Region' (e.g., 'Cloud Tech-Texas'), we'll split it into 'Department' and 'Region' columns.

```python
if 'Department_Region' in df.columns:
    # Split 'Department_Region' into 'Department' and 'Region'
    df[['Department', 'Region']] = df['Department_Region'].str.split('-', expand=True)
    print("\nDataFrame after splitting 'Department_Region':")
    display(df[['Department_Region', 'Department', 'Region']].head())
    # Optionally, drop the original compound column
    df.drop(columns=['Department_Region'], inplace=True)
else:
    print("No 'Department_Region' column found for splitting.")
display(df.head())
```

### Inconsistent Department Names
We'll standardize department names by converting them to a consistent case and stripping whitespace. We can also map common variations to a single name if needed (e.g., 'IT' and 'Information Technology').

```python
if 'Department' in df.columns:
    print("Original unique Department names:")
    display(df['Department'].unique())

    # Standardize department names (e.g., strip whitespace, title case)
    df['Department'] = df['Department'].str.strip().str.title()

    # Example: Mapping inconsistent names if they exist
    # department_mapping = {
    #     'It': 'Information Technology',
    #     'Hr': 'Human Resources'
    # }
    # df['Department'] = df['Department'].replace(department_mapping)

    print("\nCleaned unique Department names:")
    display(df['Department'].unique())
else:
    print("No 'Department' column found to standardize names.")
display(df.head())
```

### Final Cleaned Dataset
Here's the information and a preview of the dataset after all cleaning and transformation steps.

```python
print("Final Cleaned DataFrame Info:")
df.info()
print("\nFinal Cleaned DataFrame Head:")
display(df.head())
```
