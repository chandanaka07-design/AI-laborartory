```python
import numpy as np

#creating matrices directly using numpy
matrix_a = np.array([[10, 20], [30, 40]])
matrix_b = np.array([[5, 5], [10, 10]])

#matrix addition
addition = matrix_a + matrix_b

#matrix multiplication
elementwise_mult = matrix_a*matrix_b

#find max and min
max_val = np.max(matrix_a)
min_val = np.min(matrix_a)

print("addition:\n", addition)
print("multiplication:\n", elementwise_mult)
print("maximum in matrix a", max_val)
print("minimum in matrix a", min_val)
```
