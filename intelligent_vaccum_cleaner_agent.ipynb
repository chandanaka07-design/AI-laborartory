```python
rooms = [
    [1,0,],
    [0,2,],
]

rows = len(rooms)
cols = len(rooms[0])

cleaned_rooms = 0
print("initial room status:")
for row in rooms:
    print(row)

print("\nVaccum cleaner started...\n")

for i in range(rows):
  for j in range(cols):
    print("vaccum cleaner at room ({}, {})".format(i, j))

    if rooms[i][j] == 1:
      print("room is dirty")
      print("cleaning the room..")
      rooms[i][j] = 0
      cleaned_rooms += 1
      print("room cleaned")
    else:
      print("room is already clean")
    print("-"*30)

print("\nfinal room status:")
for row in rooms:
  print(row)


print("\nTotal dirty rooms cleaned=", cleaned_rooms)
print("all rooms are clean")
```
