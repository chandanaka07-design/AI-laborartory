```python
# Monkey banana problem

#Initial positions
monkey_position="door"
box_position="window"
banana_position="center"

has_banana=False

print("initial state")
print("monkey position:",monkey_position)
print("box position:",box_position)
print("banana position:",banana_position)
print()

#step 1: monkey moves to box
print("step 1: monkey walks to the box.")
monkey_position=box_position

#step 2: monkey pushes box to banana
print("step 2:monkey pushes the box under the banana.")
box_position=banana_position
monkey_position=banana_position

#step 3: monkey climbs the box
print("step 3:monkey climbs the box.")

#step 4: monkey grabs banana
print("step 4:monkey grabs the banana.")
has_banana=True

print("\nFinal state")
if has_banana:
  print("goal achived!")
  print("monkey position:",monkey_position)
  print("box position:",box_position)
  print("banana position:",banana_position)
else:
  print("goal not achieved,")
```
